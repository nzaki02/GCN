Taking all the unique nodes in protein-protein network

```
L = []
f = open("/content/mouse_interaction_data/RNA_RNA_interaction/e_RNAInter_mouse_RRI.txt","r")
count = 0
for line in f:
  if line.split()[0] not in L:
     L.append(line.split()[0])
  if line.split()[1] not in L:
     L.append(line.split()[1])
  count = count + 1
  print(count)
```

```
len(L)
```

For complexes additional ids

```
f = open("/content/reference_complexes/CORUM-Mouse-Complexes_r.txt","r")
for line in f:
  for i in line.split():
    if i not in L:
      L.append(i)
    print(count)
    count = count + 1
```

```
len(L)
```

```
G = {}
count = 0
for i in L:
  G[i] = count
  count = count +1 
```

writing complexes

```
f = open("/content/reference_complexes/CORUM-Mouse-Complexes_r.txt","r")
g = open('/content/mouse-complexes.txt',"w")
for line in f:
  for i in line.split():
    g.write(str(G[i])+'\t')
  g.write('\n')
g.close()
f.close()
```

writing ppi

```
f = open("/content/mouse_interaction_data/RNA_RNA_interaction/e_RNAInter_mouse_RRI.txt","r")
g = open('/content/mouse-ppi.txt',"w")
for line in f:
  for i in line.split()[0:2]:
    g.write(str(G[i])+'\t')
  g.write('\n')
g.close()
f.close()
```
